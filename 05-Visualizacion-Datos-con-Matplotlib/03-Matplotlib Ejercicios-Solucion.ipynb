# Ejercicios

Siga las instrucciones para recrear las gráficas con estos datos:

## Datos

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

**Importe matplotlib.pyplot como plt y configure % matplotlib inline si está utilizando el cuaderno jupyter. ¿Qué comando usa si no está usando el cuaderno jupyter?**

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Ejercicio 1

**Siga estos pasos:**

* ** Cree un objeto de figura llamado fig usando plt.figure () **
* ** Use add_axes para agregar un eje al lienzo de la figura en [0,0,1,1]. Llame a este nuevo eje ax. **
* ** Grafique (x, y) en esos ejes y establezca las etiquetas y títulos para que coincidan con el gráfico a continuación: **

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('titulo')
```

## Ejercicio 2
** Crea un objeto de figura y ponle dos ejes, ax1 y ax2. Ubicado en [0,0,1,1] y [0.2,0.5,.4,.4] respectivamente. **

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.2, 0.5, .4, .4])

axes1.plot()
axes1.set_xlim([0, 1])
axes1.set_ylim([0, 1])
axes2.plot()
axes2.set_xlim([0, 1])
axes2.set_ylim([0, 1])
```

** Ahora grafique (x, y) en ambos ejes. Y llama a tu objeto figura para mostrarlo. **

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.2, 0.5, .4, .4])

axes1.plot(x, y)
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes2.plot(x, y)
axes2.set_xlabel('x')
axes2.set_ylabel('y')
```

## Ejercicio 3

** Cree la gráfica de abajo agregando dos ejes a un objeto figura en [0,0,1,1] y [0.2,0.5, .4, .4] **

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.2, 0.5, .4, .4])

axes1.plot()
axes1.set_xlim([0, 1])
axes1.set_ylim([0, 1])
axes2.plot()
axes2.set_xlim([0, 1])
axes2.set_ylim([0, 1])
```

** Ahora use las matrices **x**, **y** y **z** para recrear el diagrama a continuación. Observe los xlimits y los límites y en la gráfica insertada: **

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.2, 0.5, .4, .4])

axes1.plot(x, y)
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes2.plot(x, y, z, color = "#FF8C00")
axes2.set_xlim([20, 22])
axes2.set_ylim([30, 50])
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('zoom')
```

## Ejercicio 4

** Utilice plt.subplots(nrows = 1, ncols = 2) para crear el gráfico a continuación. **

```python
fig, axes =plt.subplots(nrows = 1, ncols = 2)
```

** Ahora grafique (x,y) y (x,z) en los ejes. Juega con el ancho de línea y el estilo **

```python
axes[0].plot(x, y, 'b--', lw = 3)
axes[1].plot(x, z, 'r', lw = 3)
fig
```

** Ve si puedes cambiar el tamaño de la gráfica agregando el argumento figsize() en plt.subplots() (copiando y pegando tu código anterior). **

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (12, 2))
axes[0].plot(x, y, 'b', lw = 3)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[1].plot(x, z, 'r--', lw = 3)
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
